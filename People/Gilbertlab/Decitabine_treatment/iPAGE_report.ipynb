# iPAGE
https://tavazoielab.c2b2.columbia.edu/iPAGE/

iPAGE takes two column tab seprated file which first column is the gene name/id and second column are:
- RNA-seq: log2(Fold Change)
- Screening: gamma-score
- RiboSeq: ln(TE)
- meRIPSeq: differential methylation
